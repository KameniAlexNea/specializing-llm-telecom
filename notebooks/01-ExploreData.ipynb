```python
training = json.load(open("data/zindi_data/TeleQnA_training.json"))
testing = json.load(open("data/zindi_data/TeleQnA_testing1.json"))
extra_dataset = json.load(open("data/TeleQnA.json"))


len(training), len(testing), len(extra_dataset)
```

```python
from datasets import Dataset
```

```python
def patch_raw(raw: dict):
    for i in range(2, 6):
        raw[f"option {i}"] = raw.get(f"option {i}")
    return raw
training_pashed = [
    patch_raw(raw) for raw in training.values()
]
training_pashed = Dataset.from_list(training_pashed)
training_pashed
```

```python
testing_pashed = [
    patch_raw(raw) for raw in testing.values()
]
testing_pashed = Dataset.from_list(testing_pashed)
testing_pashed
```

```python
extra_pashed = [
    patch_raw(raw) for raw in extra_dataset.values()
]
extra_pashed = Dataset.from_list(extra_pashed)
extra_pashed
```
